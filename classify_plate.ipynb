```python
import os
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt

# Load your trained model once
from tensorflow.keras.models import load_model

model = tf.keras.models.load_model("C:\RoadSafeAi\models\plate_classifier.h5")

# Adjust image size based on your model's input
IMG_SIZE = (64, 128)  # Update if different

def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.resize(image, IMG_SIZE)
    image = image / 255.0  # Normalize
    return np.expand_dims(image, axis=0)

def classify_plate(image_path, visualize=False, threshold=0.4):
    img_input = preprocess_image(image_path)
    prediction = model.predict(img_input)[0][0]

    label = "RTO regulation" if prediction >= threshold else "Not RTO regulation"

    if visualize:
        img_display = cv2.imread(image_path)
        plt.imshow(cv2.cvtColor(img_display, cv2.COLOR_BGR2RGB))
        plt.title(f"{label} ({prediction:.2f})")
        plt.axis("off")
        plt.show()

    return label, prediction
```
